7.Mortgage Rate Enrichment 

-Objective 

Fetch the FRED MORTGAGE30US series, resample it from weekly to monthly frequency, and merge it onto both combined datasets using a year-month key derived from transaction dates.

1）.I fetched the MORTGAGE30US dataset from FRED using a CSV link and converted the weekly data into monthly averages to match the housing data frequency.

2）.Next, I created a year_month key for both datasets using CloseDate for the sold data and ListingContractDate for the listings data. This allowed both datasets to align with the monthly mortgage rate data.


| year_month  |rate_30yr_fixed|
|-------------|---------------|
| 1971-04    |       7.3100|
| 1971-05     |      7.4250|
|  1971-06     |      7.5300|
| 1971-07     |      7.6040|
| 1971-08     |      7.6975|

3）.Then, I merged the monthly mortgage rate data into both datasets using a left join on the year_month column. This added a new feature, rate_30yr_fixed, to each record.

4）.Validation

After merging, I checked for missing values in the mortgage rate column to ensure that the join was successful. I also previewed the merged datasets to confirm that the mortgage rate values were correctly aligned with each transaction date.

This enrichment step allows for further analysis of the relationship between mortgage rates and housing market trends. For example, it enables future analysis on whether higher mortgage rates are associated with lower home prices or longer days on market.
