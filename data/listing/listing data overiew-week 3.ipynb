```python
# =========================
# Import
# =========================
import pandas as pd
from IPython.display import display

# =========================
# Step 1: Read merged CSV
# =========================
# This dataset is already concatenated (NO filtering applied yet)
df = pd.read_csv("CRMLSListing_merged.csv", low_memory=False)

# =========================
# Step 2: Append Check (验证合并后的数据)
# =========================
print("=== APPEND CHECK ===")
print("Row count of combined/appended dataset:", df.shape[0])

# =========================
# Step 3: PropertyType BEFORE filtering
# =========================
print("\n=== PropertyType BEFORE filter ===")
prop_before = df["PropertyType"].value_counts(dropna=False)
display(prop_before)

# =========================
# Step 4: Row count BEFORE Residential filter
# =========================
rows_before_filter = df.shape[0]

print("\n=== FILTER CHECK ===")
print("Rows BEFORE Residential filter:", rows_before_filter)
```

```python
# Step 5: Apply Residential filter
# =========================
# Keep only Residential properties
df_res = df[df["PropertyType"].fillna("").str.upper() == "RESIDENTIAL"].copy()

# =========================
# Step 6: Row count AFTER Residential filter
# =========================
rows_after_filter = df_res.shape[0]
print("Rows AFTER Residential filter:", rows_after_filter)

# =========================
# Step 7: PropertyType AFTER filtering
# =========================
print("\n=== PropertyType AFTER filter ===")
prop_after = df_res["PropertyType"].value_counts(dropna=False)
display(prop_after)
```

```python
# Step 8: Validation Summary Table（加分项🔥）
# =========================
validation_summary = pd.DataFrame({
    "Check": [
        "Combined dataset row count (after concatenation)",
        "Rows BEFORE Residential filter",
        "Rows AFTER Residential filter"
    ],
    "Value": [
        df.shape[0],
        rows_before_filter,
        rows_after_filter
    ]
})

print("\n=== VALIDATION SUMMARY ===")
display(validation_summary)
```

```python
# Step 9: Save filtered dataset
# =========================
df_res.to_csv("CRMLSListing_residential.csv", index=False)

print("\n✅ File saved: CRMLSSold_residential.csv")
```
