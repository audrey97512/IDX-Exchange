```python
# =========================
# Import
# =========================
from pathlib import Path
import pandas as pd

# =========================
# Step 1: 设置正确路径
# =========================
folder = Path(".")   # 当前文件夹

# 找所有 Listing CSV
csv_files = sorted(folder.glob("CRMLSListing*.csv"))

# =========================
# Step 2: Debug（确认读到文件）
# =========================
print("=== Files Found ===")
print("Number of files:", len(csv_files))

for f in csv_files:
    print(f.name)

# =========================
# Step 3: 初始化
# =========================
summary = []
df_list = []

# =========================
# Step 4: 读取 + 统计
# =========================
for file in csv_files:
    try:
        df = pd.read_csv(file, encoding="utf-8", low_memory=False)
        used_encoding = "utf-8"
    except UnicodeDecodeError:
        df = pd.read_csv(file, encoding="latin1", low_memory=False)
        used_encoding = "latin1"

    print(f"{file.name}: {df.shape[0]} rows")

    summary.append({
        "file": file.name,
        "num_rows": df.shape[0],
        "num_columns": df.shape[1],
        "encoding_used": used_encoding
    })

    df["source_file"] = file.name
    df_list.append(df)

# =========================
# Step 5: 生成 summary 表
# =========================
summary_df = pd.DataFrame(summary)

print("\n=== Summary Table ===")
print(summary_df)

# =========================
#  # Step 6: 合并 + 导出 CSV
# =========================
if len(df_list) == 0:
    print("❌ ERROR: No data loaded. Check your path!")
else:
    combined_df = pd.concat(df_list, ignore_index=True)
    df = combined_df

    print("\n=== Combined Data ===")
    print("Total rows:", df.shape[0])
    print("Total columns:", df.shape[1])

   
   
```

```python
 # ✅ 导出为一个CSV文件
df.to_csv("CRMLSListing_merged.csv", index=False)
```
